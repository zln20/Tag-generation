```python
import pandas as pd
import os
import re
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

# Train_docs

```python
train_doc=r"C:\Users\samya\Downloads\Lawnics_Assignment_NLP\Train_docs"
train_docs=os.listdir(train_doc)
```

```python
train_text=[]
for i in train_docs:
    t=open(os.path.join(train_doc,i))
    text=t.read()
    t.close()
    text=text.replace('\n','')
    train_text.append(text)
```

```python
train_text=pd.DataFrame(train_text,columns=["text"])
train_text
```

# Train_tags

```python
train_tag=r"C:\Users\samya\Downloads\Lawnics_Assignment_NLP\Train_tags"
tags=os.listdir(train_tag)
```

```python
train_tags=[]
for i in tags:
    t=open(os.path.join(train_tag,i))
    tag=t.read()
    t.close()
    tag=tag.replace('\n','')
    train_tags.append(tag)
```

```python
train_tags=pd.DataFrame(train_tags,columns=["tags"])
train_tags
```

# Test_text

```python
test_doc=r"C:\Users\samya\Downloads\Lawnics_Assignment_NLP\Test_docs"
test_docs=os.listdir(test_doc)
```

```python
test_text=[]
for i in test_docs:
    t=open(os.path.join(test_doc,i))
    text=t.read()
    t.close()
    text=text.replace('\n','')
    test_text.append(text)
```

```python
test_text=pd.DataFrame(test_text,columns=["text"])
test_text
```

# Proprocessing

## Filtering 

```python
ps=PorterStemmer()
train_t=[]
for i in range(0,len(train_text)):
    token=re.sub('[^a-zA-Z]',' ',str(train_text["text"][i]))
    token=token.lower()
    token=token.split()
    token=[ps.stem(word) for word in token if not word in stopwords.words('english')]
    token=' '.join(token)
    train_t.append(token)
```

```python
test_t=[]
for i in range(0,len(test_text)):
    token=re.sub('[^a-zA-Z]',' ',str(test_text["text"][i]))
    token=token.lower()
    token=token.split()
    token=[ps.stem(word) for word in token if not word in stopwords.words('english')]
    token=' '.join(token)
    test_t.append(token)
```

```python
labels = list()
for item in train_tags.tags:
    temp = item.split(',')
    for x in temp:
        x_ = x.strip()
        if x_ not in labels:
            labels.append(x_) 
    else:
            continue
labels
```

# Count vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

x=cv.fit_transform(train_t).toarray()
test=cv.fit_transform(test_t).toarray()

x=pd.DataFrame(x)
test=pd.DataFrame(test)
```

```python
y=pd.DataFrame()

for index, row in train_tags.iterrows():
    for label in labels:
        if label in row['tags']:
            y.at[index, label] =1
y=y.fillna(0)
y
```

# Train test split

```python
missing_cols = set( test.columns ) - set( x.columns )
for c in missing_cols:
    x[c] = 0

x = x[test.columns]
```

```python
from sklearn.model_selection import train_test_split as tts
train_x,test_x,train_y,test_y=tts(x,y,test_size=0.1,random_state=20)
```
